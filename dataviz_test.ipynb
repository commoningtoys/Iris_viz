```python
import pandas as pd
import numpy as np
import random
```

```python
rawdataDF = pd.read_json('RAW_DATA.json')
```

```python
vizDF = pd.DataFrame()
fld = []
resting = []
working = []
#trading = []
colors = [('cook','red'),('clean','blue'),('admin','green'),('shop','yellow')]
for n in range(len(rawdataDF)):
    for agent in rawdataDF.columns:
        fld.append(rawdataDF[agent][n]['feel_like_doing'])
        resting.append(rawdataDF[agent][n]['feel_like_doing'])
        #check if trading; if yes, that will be its position
        if rawdataDF[agent][n]['traded'] == '':
            #get the position of the agent based on 'executed_task'; color by preferences>task_preference>maximum value
            if rawdataDF[agent][n]['executed_task'] == 'cook':
                x_pos=random.randint(100,500)
                y_pos=random.randint(100,300)
                color = colors[next(iter([j for j,x in enumerate([rawdataDF[agent][n]['prefereces'][item]['task_preference'] for item in ['cook','clean','admin','shop']]) if x == max([rawdataDF[agent][n]['prefereces'][item]['task_preference'] for item in ['cook','clean','admin','shop']])]))][1]
            elif rawdataDF[agent][n]['executed_task'] == 'clean':
                x_pos=random.randint(700,1100)
                y_pos=random.randint(100,300)
                color = colors[next(iter([j for j,x in enumerate([rawdataDF[agent][n]['prefereces'][item]['task_preference'] for item in ['cook','clean','admin','shop']]) if x == max([rawdataDF[agent][n]['prefereces'][item]['task_preference'] for item in ['cook','clean','admin','shop']])]))][1]
            elif rawdataDF[agent][n]['executed_task'] == 'admin':
                x_pos=random.randint(100,500)
                y_pos=random.randint(500,700)
                color = colors[next(iter([j for j,x in enumerate([rawdataDF[agent][n]['prefereces'][item]['task_preference'] for item in ['cook','clean','admin','shop']]) if x == max([rawdataDF[agent][n]['prefereces'][item]['task_preference'] for item in ['cook','clean','admin','shop']])]))][1]
            elif rawdataDF[agent][n]['executed_task'] == 'shop':
                x_pos=random.randint(700,1100)
                y_pos=random.randint(500,700)
                color = colors[next(iter([j for j,x in enumerate([rawdataDF[agent][n]['prefereces'][item]['task_preference'] for item in ['cook','clean','admin','shop']]) if x == max([rawdataDF[agent][n]['prefereces'][item]['task_preference'] for item in ['cook','clean','admin','shop']])]))][1]        
            else:
                print('agent has not executed any tasks')
        else:
           #print(rawdataDF[agent][i]['traded'])
            x_pos=random.randint(100,1100)
            y_pos=random.randint(850,950)
            color = colors[next(iter([j for j,x in enumerate([rawdataDF[agent][n]['prefereces'][item]['task_preference'] for item in ['cook','clean','admin','shop']]) if x == max([rawdataDF[agent][n]['prefereces'][item]['task_preference'] for item in ['cook','clean','admin','shop']])]))][1]
        working.append((x_pos,y_pos,color))
vizDF['fld'] = fld
vizDF['resting'] = resting
vizDF['working'] = working
```

```python
agentDF=pd.DataFrame()
n=12 #iteration
fld = []
resting = []
working = []
trading = []
colors = [('cook','red'),('clean','blue'),('admin','green'),('shop','yellow')]
for agent in rawdataDF.columns:
    fld.append(rawdataDF[agent][n]['feel_like_doing'])
    resting.append(rawdataDF[agent][n]['feel_like_doing'])
    if rawdataDF[agent][n]['executed_task'] == 'cook':
        x_pos=random.randint(100,500)
        y_pos=random.randint(100,300)
        color = colors[next(iter([j for j,x in enumerate([rawdataDF[agent][n]['prefereces'][item]['task_preference'] for item in ['cook','clean','admin','shop']]) if x == max([rawdataDF[agent][n]['prefereces'][item]['task_preference'] for item in ['cook','clean','admin','shop']])]))][1]
    working.append((x_pos,y_pos,color))
    if rawdataDF[agent][n]['traded'] == '':
        #print('WORKING')
        trading.append('none')
    else:
       #print(rawdataDF[agent][i]['traded'])
        x_pos=random.randint(50,1150)
        y_pos=random.randint(900,950)
        trading.append((x_pos,y_pos,rawdataDF[agent][n]['traded']))
agentDF['fld'] = fld
agentDF['resting'] = resting
agentDF['working'] = working
agentDF['trading'] = trading
```

```python
from bokeh.io import push_notebook, show, output_notebook, export_png
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, LabelSet
output_notebook()
```

```python
#plot data from one iteration
x = np.array([line[0] for line in agentDF['working']])
y = np.array([line[1] for line in agentDF['working']])
radii = np.interp(np.array(agentDF['fld']), (np.array(agentDF['fld']).min(), np.array(agentDF['fld']).max()), (0, 20))
colors = [line[2] for line in agentDF['working']]

TOOLS = "crosshair,pan,wheel_zoom,box_zoom,reset,box_select,lasso_select"

# create a new plot with the tools above, and explicit ranges
p = figure(tools=TOOLS, x_range=(0, 1200), y_range=(0, 1000))

# add a circle renderer with vectorized colors and sizes
p.circle(x, y, radius=radii, fill_color=colors, fill_alpha=0.6, line_color=None)

# show the results
#export_png(p, filename="plot.png") 
show(p)
```
